  b. Create data set “cust_rightjoin” using right Join

```python
clj=pd.merge(data,demo,on="ID",how="left")
crj=pd.merge(data,demo,on="ID",how="right")
cij=pd.merge(data,demo,on="ID",how="inner")
cfj=pd.merge(data,demo,on="ID",how="outer")
```
