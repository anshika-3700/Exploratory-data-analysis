### 1. Compare triglycerides and weights before and after of same people and conclude whether change is effective using hypothesis testing through t-statistics and p-value

```python
diet=pd.read_csv("dietstudy.csv")
```

```python
diet.head()
```

```python
diet["tg"]=diet[["tg1","tg2","tg3","tg4"]].mean(axis=1)
```

```python
diet["wg"]=diet[["wgt1","wgt2","wgt3","wgt4"]].mean(axis=1)
```

```python
diet.head()
```

```python
# Example of the Paired Student's t-test
# assuming confidence interval of 95% , so significance level (a=.05)
# H0= no significant change in values of triglycerides of person after a certain interval 
# HA=!H0

from scipy.stats import ttest_rel
t , p = ttest_rel(diet.tg0,diet.tg)
print('t-score =%.2f, p-value =%.2f' % (t, p))
if p > 0.05:
    print('Statistically the same distribution, accepting H0')
else:
    print('Statistically different distributions, reject H0')
```

```python
# Assuming confidence interval of 95% , so significance level (a=.05)
# H0= no significant change in values of weight of person after a certain interval 
# HA=!H0

tw , pw = ttest_rel(diet.wgt0,diet.wg)
print('t-score =%.2f, p-value =%.4f' % (tw, pw))
if p > 0.05:
    print('Probably the same distribution,accepting H0')
else:
    print('Probably different distributions,reject H0')
```

### 2. Compare sales from two groups of people - promotional ad and standard . Conclude whether output of promotional activity is different from standard or not.

```python
credit=pd.read_csv("creditpromo.csv")
```

```python
credit.head()
```

```python
st=credit[credit["insert"]=="Standard"]
np=credit[credit["insert"]=="New Promotion"]
round(st.dollars.mean(),2),round(np.dollars.mean(),2)
```

```python
# independent t-test
# Assuming confidence interval of 95% , so significance level (a=.05)
# H0= no significant difference in both the campaign activity 
# HA= new promotion activity was effective in generating more sales than standard

from scipy.stats import ttest_ind
t,p=ttest_ind(st.dollars,np.dollars)
print('t-score =%.2f, p-value =%.2f' % (t, p))
if p > 0.05:
    print('Accepting H0')
else:
    print('Reject H0')
```

### 3. Compare hybrid seed production from two different methods of pollination 

```python
pol=pd.read_csv("pollination.csv")
```

```python
pol
```

### a. Is the overall population of Seed yield/plant (g) equals to 200?

```python
# Assuming confidence interval of 95% , so significance level (a=.05)
# H0 (null hypothesis)= average of Seed_Yield_Plant of all plants is equal to 200g
# HA(alt hypothesis)=!H0

from scipy.stats import ttest_1samp
t,p=ttest_1samp(pol.Seed_Yield_Plant,200)
print('t-score =%.2f, p-value =%.3f' % (t, p))
if p > 0.05:
    print('Accepting H0')
else:
    print('Reject H0')
```

### b. Test whether the natural pollination and hand pollination under open field conditions are equally effective or are significantly different

```python
nat=pol[pol["Group"]=="Natural"]
hd=pol[pol["Group"]=="Hand"]
```

```python
nat=nat.drop("Group",axis=1)
hd=hd.drop("Group",axis=1)
```

```python
nat.head(2)
```

```python
# Example of the Analysis of Variance Test
from scipy.stats import f_oneway
t,p=f_oneway(nat,hd)
print('stat=%.3f, p=%.3f' % (t, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

### 4. Compare dvd rating by different age-groups , whether different or same .

```python
dvd=pd.read_csv("dvdplayer.csv")
dvd.head()
```

```python
dvd.agegroup.value_counts()
```

```python
dvd.groupby("agegroup").mean().round(3)
```

```python
a=dvd.loc[dvd.agegroup=="25-34"].reset_index().drop(["index","agegroup"],axis=1)
b=dvd.loc[dvd.agegroup=="35-44"].reset_index().drop(["index","agegroup"],axis=1)
c=dvd.loc[dvd.agegroup=="45-54"].reset_index().drop(["index","agegroup"],axis=1)
d=dvd.loc[dvd.agegroup=="55-64"].reset_index().drop(["index","agegroup"],axis=1)
e=dvd.loc[dvd.agegroup=="55-64"].reset_index().drop(["index","agegroup"],axis=1)
f=dvd.loc[dvd.agegroup=="55-64"].reset_index().drop(["index","agegroup"],axis=1)
```

```python
a,b
```

```python
# Assuming confidence interval of 95% , so significance level (a=.05)
# H0 (null hypothesis)= ratings by a,b,c,d,e,f are almost same
# HA(alt hypothesis)=!H0

from scipy.stats import f_oneway 
t,p=f_oneway(a,b,c,d,e,f)
print('stat=%.3f, p=%.3f' % (t, p))
if p > 0.05:
    print('Accepting H0')
else:
    print('Reject H0')
```

### a. Is there any relationship in between labour force status with marital status?

```python
ss1=pd.crosstab(ss.wrkstat, ss.marital,margins = True)
ss1
```

```python
from scipy.stats import chi2_contingency
x2 = chi2_contingency(ss1)
x2
```

```python
# Assuming confidence interval of 95% , so significance level (a=.05)
# H0 (null hypothesis)= work status and marital status are dependent or associated with each other
# HA(alt hypothesis)=!H0

print('stat=%.3f, p=%.3f' % (x2[0],x2[1]))
if p > 0.05:
    print('Accepting H0')
else:
    print('Reject H0')
```
